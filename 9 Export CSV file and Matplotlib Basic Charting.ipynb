```python
%matplotlib inline
# it makes chart inside the jupyter so its good feature
# import matplotlib
# Dont import whole matplotlib because we need some only.
from matplotlib import pyplot as plt 
# plt.show() does not work as its open in browser.
plt.plot([1,2,3],[2,5,7])
plt.show()
```

```python
# show headings and x y labels
plt.plot([1,2,3],[2,5,7])
plt.title("Sahil Data")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```

```python
# ek hi plot me 2 line/equation show krni h for comparision.
# for style 
from matplotlib import style
style.use('ggplot')
# search matplotlib style in google
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y,label='first',linewidth=8)
plt.plot(x2,y2,label='second')
plt.legend()
plt.show()
```

```python
# for scatter plot
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.scatter(x,y,label='first',linewidth=3)
plt.scatter(x2,y2,label='second')
plt.legend()
plt.show()
```

```python
# for histogram
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.hist(x,label='first',linewidth=8)
plt.hist(x2,label='second')
plt.legend()
plt.show()
```
